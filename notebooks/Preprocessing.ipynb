```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open("../assets/data/samples/sample_20.JPEG")
image = np.array(image)
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(gray_image.shape)
plt.imshow(gray_image, cmap="gray")
```

```python
circles = cv2.HoughCircles(gray_image, method=cv2.HOUGH_GRADIENT, dp=1, minDist=10, param1=10, param2=30, minRadius=30, maxRadius=70)
```

```python
print(circles)
```

```python
circled_image = image
for circle in circles[0]:
    x, y, radius = circle
    circled_image = cv2.circle(circled_image, (x, y), radius, (0,0,255),2)
plt.imshow(circled_image)
```
